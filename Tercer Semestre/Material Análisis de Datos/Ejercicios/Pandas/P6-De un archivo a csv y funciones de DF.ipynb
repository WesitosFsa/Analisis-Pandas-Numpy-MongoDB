```python
import pandas as pd
convertir = pd.read_json('archivojson.json')
convertir.to_csv('archivojson.csv', index=None, header=True)
```

```python
#tiene el encoding por que tiene ñ 
df = pd.read_csv('archivojson.csv', encoding='latin-1')
df
```

```python
#acceso mediante cordenada de posicion
# filas[i ] columnas [j]
df.iloc[1,0]
```

```python
#mediante un rango
#nos va a soltar todos los datos hasta el rango de la columna menos la zona
df.iloc[1,:3]
```

```python
#acceso de los elementos mediante un nombre
df.loc[2,'GENERO']
#acceso a mas de 2 elementos 
df.loc[:2,('GENERO','CARRERA')]
```

```python
#añadir columnas a un dataframe
df['EXTRA']= pd.Series([1,2,3,4,5])
df
```

```python
#eliminar columna de un dataframe
EXTRA = df.pop('EXTRA')
df
```

```python
# Generar una lista con los datos que se quieren agregar
data = [['Carlos', 27, 'M']]
# Crear un nuevo DataFrame a partir de la lista data
# Especificar las columnas en el nuevo DataFrame usando el argumento columns
nueva_fila = pd.DataFrame(data, columns=['NOMBRE', 'EDAD', 'GENERO'])
# El argumento ignore_index=True se utiliza para que los índices se vuelvan a numerar de forma continua después de la concatenación
df = pd.concat([df, nueva_fila], ignore_index=True)
df
```

```python
# Eliminar filas del DataFrame
df = df.drop([8])
# Imprimir el DataFrame resultante
df
```

```python
#FILTRAR PARA BUSCAR DATOS
df[(df['GENERO']=='F') & (df['EDAD']>=22)]
```

```python
#ORDENAR EL DATAFRAME DE MANERA ALFABETICA
df = df.sort_values('CARRERA')
df
```

```python
#ORDENAR EL DATAFRAME DE MANERA NUMERICA
df = df.sort_values('EDAD')
df
```
